```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# file_path = '/Users/adrian/Documents/UV-vis/Carlos_P3HT_8-22-24/Polarized Batch 4 0% Samples  - Horizontal 1.csv'
file_path = '/Users/adrian/Documents/UV-vis/Carlos_P3HT_8-22-24/Polarized Batch 4 0% Samples  - Vertical 1.csv'
df = pd.read_csv(file_path)

# baseline_file_path = "/Users/adrian/Documents/UV-vis/baselines/Polarized Batch 4 0% Samples  - Baseline Horizontal.csv"
baseline_file_path = "/Users/adrian/Documents/UV-vis/baselines/Polarized Batch 4 0% Samples  - Baseline Vertical.csv"
baseline_df = pd.read_csv(baseline_file_path)

wavelength = df['Wavelength']
absorbance_data = df.drop(columns=['Wavelength'])
absorbance_data = absorbance_data.drop(columns=['100%']) # drop 100% because the PDMS get to narrow to completely occlude beam

baseline_absorbance_data = baseline_df['Average'].values[:, None]  # Convert to a 2D array for broadcasting
corrected_absorbance = absorbance_data - baseline_absorbance_data
max0 = max(corrected_absorbance['0%'])
corrected_absorbance = corrected_absorbance/max0

#plotting 
plt.figure(figsize=(10, 6))
colors = cm.Spectral([i / (absorbance_data.shape[1]) for i in range(absorbance_data.shape[1])])

# Plot the corrected absorbance data with solid lines
for i, column in enumerate(corrected_absorbance.columns): 
    plt.plot(wavelength, corrected_absorbance[column], label=f'{column}', color=colors[i])

# Plot the baseline data in black
# plt.plot(wavelength, baseline_absorbance_data, label='||-Baseline', color='black', linewidth=2)

plt.title('Normalized Absorbance vs. Wavelength for ||-polarized light')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorbance')
plt.xlim((400,700))
plt.ylim((0,1.05))
plt.legend(title='Strain', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
```
